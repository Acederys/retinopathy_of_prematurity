```python
import numpy as np
import os
import pandas as pd
import matplotlib.pyplot as plt
import cv2
from pathlib import Path

data_R = '/content/drive/MyDrive/2_semestr/PP/Ретинопатия_1/have'
data_N = '/content/drive/MyDrive/2_semestr/PP/Ретинопатия_1/not'
```

```python
files_R = os.listdir(data_R)
files_N = os.listdir(data_N)
```

```python
df = pd.DataFrame({'image': files_R, 'type': 1})
df_1 = pd.DataFrame({'image': files_N, 'type': 0})
```

```python
DF = pd.concat([df, df_1], ignore_index=True)
```

```python
DF.to_csv('R_classification.csv')
```

```python
df_2 = pd.DataFrame(pd.concat([DF.loc[DF['image'].str.contains(r'_2-3_')],DF.loc[DF['image'].str.contains(r'N_')]]))
```

```python
df_2
```

```python
df_2.loc[df_2['image'].str.contains(r'_2-3_'),'type'] = 2
df_2.loc[df_2['image'].str.contains(r'N_'),'type'] = 0
df_2.loc[df_2['image'].str.contains(r'_PLUS'),'type'] = 'PLUS'
```
