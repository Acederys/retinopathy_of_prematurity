```python
import matplotlib.pyplot as plt
import zipfile
from google.colab import drive
from PIL import Image
import os
```

```python
drive.mount('/content/drive')
```

```python
zip_file_path = '/content/drive/MyDrive/ROP_datasets/second_dataset.zip'
# Распаковать архив в текущую папку
with zipfile.ZipFile(zip_file_path, 'r') as zip_ref:
    zip_ref.extractall()
```

```python
i = 0
for file_name1, file_name2, file_name3 in zip(os.listdir("./second_dataset/healthy"), os.listdir("./second_dataset/stage 2-3"), os.listdir("./second_dataset/plus")):
    plt.figure(figsize=(10, 10))
    i += 1

    img = Image.open(f"./second_dataset/healthy/{file_name1}")
    plt.subplot(1,3,1)
    plt.axis("off")
    plt.title("healthy")
    plt.imshow(img)

    img = Image.open(f"./second_dataset/stage 2-3/{file_name2}")
    plt.subplot(1,3,2)
    plt.axis("off")
    plt.title("stage 2-3")
    plt.imshow(img)

    img = Image.open(f"./second_dataset/plus/{file_name3}")
    plt.subplot(1,3,3)
    plt.axis("off")
    plt.title("plus")
    plt.imshow(img)
    plt.show()
    if i == 3: break
```
