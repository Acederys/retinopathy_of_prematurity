```python
import matplotlib.pyplot as plt
import zipfile
from google.colab import drive
from PIL import Image
import os
```

```python
drive.mount('/content/drive')
```

```python
zip_file_path = '/content/drive/MyDrive/ROP_datasets/first_dataset.zip'
# Распаковать архив в текущую папку
with zipfile.ZipFile(zip_file_path, 'r') as zip_ref:
    zip_ref.extractall()
```

```python
i = 0
for file_name1, file_name2 in zip(os.listdir("./first_dataset/healthy"), os.listdir("./first_dataset/unhealthy")):
    plt.figure(figsize=(10, 10))
    i += 1

    img = Image.open(f"./first_dataset/healthy/{file_name1}")
    plt.subplot(1,2,1)
    plt.axis("off")
    plt.title("healthy")
    plt.imshow(img)

    img = Image.open(f"./first_dataset/unhealthy/{file_name2}")
    plt.subplot(1,2,2)
    plt.axis("off")
    plt.title("unhealthy")
    plt.imshow(img)
    plt.show()
    if i == 3: break
```
